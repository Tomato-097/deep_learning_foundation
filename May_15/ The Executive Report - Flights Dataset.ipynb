**Dataset:** sns.load_dataset('flights')  
**Goal:** Take raw historical flight data and polish it into a presentation-ready graphic.

```python
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset("flights")
```

**Themes & Contexts:** Apply sns.set_style('whitegrid') to give it a clean background, and sns.set_context('talk') to automatically scale up the fonts for a presentation slide.

```python
sns.set_style("whitegrid")
sns.set_context("talk")
```

**Color Palettes:** Pivot the dataset into a matrix (months vs. years) and plot a heatmap. Use sns.color_palette("Blues", as_cmap=True) so the colors intuitively map from light (few passengers) to dark (many passengers).

```python
flights_matrix = flights.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(flights_matrix, cmap=sns.color_palette("Blues", as_cmap=True))
plt.title("Flight Passengers per Month (1949 - 1960)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

**Advanced Matplotlib Formatting:** Plot a line chart of average passengers per year.

```python
yearly_avg = flights.groupby("year")["passengers"].mean().reset_index()

fig, ax = plt.subplots(figsize=(12, 6))

sns.lineplot(data=yearly_avg, x="year", y="passengers", marker="o", ax=ax)

ax.set_title("Average Annual Flight Passengers", pad=20)
ax.set_xlabel("Year")
ax.set_ylabel("Average Passengers")
plt.show()
```

Rotate the X-axis labels (years) by 45 degrees so they are perfectly legible.

```python
ax.set_xticks(yearly_avg["year"])
ax.set_xticklabels(yearly_avg["year"], rotation=45)

fig
```

Use ax.set_ylim(0, 500) to force the Y-axis to start at zero instead of floating.

```python
ax.set_ylim(0, 500)

fig
```

Use ax.annotate() to draw an arrow pointing to the year 1960 with the text: "Introduction of Commercial Jets".

```python
val_1960 = yearly_avg.loc[yearly_avg["year"] == 1960, "passengers"].values[0]

ax.annotate("Introduction of Commercial Jets", xy=(1960, val_1960), xytext=(1953, 400),
    arrowprops=dict(facecolor="black", shrink=0.08, width=1, headwidth=8),)

fig
```

**Exporting:** Run plt.savefig('executive_flight_report.png', dpi=300, bbox_inches='tight') to save the final, high-resolution masterpiece to their computer.

```python
fig.savefig("executive_flight_report.png", dpi=300, bbox_inches="tight")
```
