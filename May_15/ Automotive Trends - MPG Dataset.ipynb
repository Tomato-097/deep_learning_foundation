**Dataset:** sns.load_dataset('mpg')   
**Goal:** Compare car performance and fuel efficiency across different global regions.

```python
import matplotlib.pyplot as plt
import seaborn as sns

mpg_df = sns.load_dataset("mpg")
```

**FacetGrid / Relplot / Catplot:** Use sns.relplot(x='weight', y='mpg', col='origin'). This will generate three separate scatterplots side-by-side (USA, Europe, Japan) so students can compare weight vs. fuel efficiency by region.

```python
g1 = sns.relplot(data=mpg_df, x="weight", y="mpg", col="origin")
g1.fig.suptitle("Weight vs. Fuel Efficiency by Region", y=1.05)
plt.show()
```

Then, use sns.catplot(kind='box', x='cylinders', y='horsepower', row='origin') to create a vertical stack of boxplots.

```python
g2 = sns.catplot(data=mpg_df, x="cylinders", y="horsepower", row="origin")
g2.fig.suptitle("Horsepower Distribution by Cylinders and Region", y=1.02)
plt.show()
```

**Joint Plots:** Use sns.jointplot(x='horsepower', y='mpg') to show a scatter plot in the center, with a histogram of horsepower on the top margin and a histogram of mpg on the right margin.

```python
g3 = sns.jointplot(data=mpg_df, x="horsepower", y="mpg")
g3.fig.suptitle("Overall Correlation: Horsepower vs. Fuel Efficiency", y=1.02)
plt.show()
```
