**Dataset:** sns.load_dataset('titanic') 

**Goal:** Understand who was on the Titanic by exploring single variables.  

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
# This loads a built-in sample dataset containing information about the passengers on the Titanic (such as age, sex, class, and survival status).

sns.set_theme(style="whitegrid")
```

**Categorical - Bar/Count:** Use sns.countplot() to show how many passengers were in First, Second, and Third class (class column).

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='class', hue='class', palette='viridis')

plt.title('Passenger Count by Class', fontsize=14)
plt.xlabel('Ticket Class', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.show()
```

**Categorical - Pie:** Use plt.pie() to show the raw proportion of male vs. female passengers (sex column), discussing why the countplot might have been easier to read.

```python
# Get the counts of male vs female passengers
sex_counts = titanic['sex'].value_counts()

# Plot the pie chart
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Male vs. Female Passengers')
plt.show()
```

**Numerical - Histograms:** Use sns.histplot() to find the age distribution (age column) of the passengers.

```python
sns.histplot(data=titanic, x='age')

plt.title('Age Distribution of Passengers', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

**Numerical - KDE:** Use sns.kdeplot() to look at the density of ticket prices (fare column) to see the massive skew toward cheaper tickets.

```python
sns.kdeplot(data=titanic, x='fare', color='crimson', fill=True)

plt.title('Density of Ticket Prices (Fare)', fontsize=14)
plt.xlabel('Fare (Ticket Price)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

**Combining Hist + KDE:** Overlay the KDE curve on top of a histogram for the age column using sns.histplot(data, x='age', kde=True) to see both the bins and the smooth trend.

```python
sns.histplot(data=titanic, x='age', kde=True, color='teal')

plt.title('Age Distribution with KDE Curve Overlay', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count / Density', fontsize=12)
plt.show()
```
