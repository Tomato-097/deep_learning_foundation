```python
# The Goal: Create a DigitalWallet class with private attributes for the 
#         account __balance and a __pin_code. Create methods like deposit(), 
#         withdraw(), and get_balance() that interact with these attributes 
#         securely. Ensure that withdraw() and get_balance() require the user 
#         to pass the correct PIN to work, and that the __balance cannot be 
#         directly accessed, modified, or set to a negative value from outside 
#         the class. 
# Concepts Tested: Encapsulation, private attributes (name mangling), 
#         data hiding, getter/setter methods, and access control.

class DigitalWallet:
    def __init__(self, initial_balance, pin_code):
        self.__balance = max(0, initial_balance)
        self.__pin_code = pin_code

    def __verify_pin(self, pin):
        return pin == self.__pin_code

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Successfully deposited ${amount}.")
        else:
            print("Error: Deposit amount must be positive.")

    def withdraw(self, amount, pin):
        if not self.__verify_pin(pin):
            print("Access Denied: Incorrect PIN.")
            return

        if amount <= 0:
            print("Error: Withdrawal must be a positive amount.")
        elif amount > self.__balance:
            print("Error: Insufficient funds.")
        else:
            self.__balance -= amount
            print(f"Successfully withdrew ${amount}.")

    def get_balance(self, pin):
        if self.__verify_pin(pin):
            return f"Current Balance: ${self.__balance}"
        else:
            return "Access Denied: Incorrect PIN."

wallet = DigitalWallet(100, "1234")

while True:
    print("\n--- Digital Wallet Menu ---")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")
    
    choice = input("Select an option (1-4): ")

    if choice == "1":
        pin = input("Enter your 4-digit PIN: ")
        print(wallet.get_balance(pin))

    elif choice == "2":
        try:
            amt = float(input("Enter deposit amount: "))
            wallet.deposit(amt)
        except ValueError:
            print("Invalid input. Please enter a number.")

    elif choice == "3":
        try:
            amt = float(input("Enter withdrawal amount: "))
            pin = input("Enter your 4-digit PIN: ")
            wallet.withdraw(amt, pin)
        except ValueError:
            print("Invalid input. Please enter a number.")

    elif choice == "4":
        print("Exiting...")
        break

    else:
        print("Invalid selection. Please try again.")
```
