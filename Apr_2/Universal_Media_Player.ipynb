```python
# The Goal: Create a base Media class with a play() method. Then, create 
#         subclasses like Song, Movie, and Podcast that inherit from Media and 
#         override the play() method to output unique processing messages 
#         (e.g., "Playing audio...", "Playing video at 1080p...", "Playing podcast..."). 
#         Finally, write a standalone function play_all(media_list) that loops through 
#         a list containing different media objects and triggers the play() method on 
#         each one, demonstrating how a single function handles different object types 
#         interchangeably. 
# Concepts Tested: Polymorphism, Method Overriding, Inheritance, Duck Typing.

class Media:
    def play(self):
        pass

class Song(Media):
    def play(self):
        print("Playing audio: High-fidelity stereo mode active.")

class Movie(Media):
    def play(self):
        print("Playing video at 1080p: Buffering high-definition stream...")

class Podcast(Media):
    def play(self):
        print("Playing podcast: Normalizing voice levels...")

def play_all(media_list):
    print(f"--- Starting Playlist ---")
    for item in media_list:
        item.play()
    print("--- End of Playlist ---\n")

my_media = [Song(), Movie(), Podcast(), Song()]

play_all(my_media)
```
