```python
# The Goal: Create a Pet class with attributes like name, hunger, energy, and 
# happiness. Create methods like feed(), play(), and sleep() that alter these 
# attributes.
# Concepts Tested: Classes, __init__, self, instance attributes, methods.

class Pet:
    def __init__(self, name):
        self.name = name
        self.hunger = 50     
        self.energy = 100     
        self.happiness = 50  

    def feed(self):
        print(f"You fed {self.name}!")
        self.hunger = max(0, self.hunger - 20)
        self.happiness = min(100, self.happiness + 5)

    def play(self):
        if self.energy < 20:
            print(f"{self.name} is too tired to play.")
        else:
            print(f"You played fetch with {self.name}!")
            self.happiness = min(100, self.happiness + 20)
            self.energy -= 20
            self.hunger = min(100, self.hunger + 10)

    def sleep(self):
        print(f"{self.name} is taking a nap...")
        self.energy = 100
        self.hunger = min(100, self.hunger + 5)

    def status(self):
        print(f"\n--- {self.name}'s Stats ---")
        print(f"Hunger: {self.hunger}/100")
        print(f"Energy: {self.energy}/100")
        print(f"Happiness: {self.happiness}/100\n")

my_pet = Pet("Buddy")
my_pet.status()
my_pet.play()
my_pet.feed()
my_pet.sleep()
my_pet.status()
```
