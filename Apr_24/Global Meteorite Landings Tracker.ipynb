```python
# The Scenario: You act as junior planetary scientists analyzing historical 
#             data of meteorite impacts and discoveries on Earth.
# Data Structures: You separate out specific columns, like the mass_grams, 
#             to see how it behaves as an isolated 1D Series compared to 
#             the broader 2D DataFrame context.
# Data I/O: You start by loading the raw data using pd.read_csv(). 
#             Later, you will export your filtered findings using 
#             .to_excel() or .to_json() for a museum curation 
#             team to use for an upcoming exhibit.

import pandas as pd

df = pd.read_csv('meteorites.csv')

print("2D DataFrame Context:")
print(df.head())

mass_gram =df['mass_grams']

print("\n Isolated 1D Series:")
print(mass_gram.head())
```

```python
# Initial Data Inspection: Before doing anything, you must run .info() 
#         and .describe() to understand the sheer volume of data and 
#         check for missing values. You will use .head() to see the 
#         layout of the columns, and use .value_counts() to see which 
#         meteorite classification types (e.g., Chondrite, Iron) are 
#         the most frequently found.
print("Dataset Preview:")
print(df.head())

print("\nData Structure and Null Count:")
df.info()

print("\nNumerical Statistics:")
print(df.describe())

print("\nFrequency of Meteorite Classifications:")
classification_counts = df['classification'].value_counts()
print(classification_counts)
```

```python
# Indexing and Selection: 
# Column/Label/Integer: Use dictionary-style selection to grab the 
#             fall_status column (whether it was seen falling or 
#             just found later). Use .iloc[] to extract the top 
#             5 heaviest meteorites based on their sorted row
#             position.

fall_status = df['fall_status']

print("--- Fall Status Column ---")
print(fall_status.head())

df_sorted = df.sort_values(by='mass_grams', ascending=False)

heaviest_meteorites = df_sorted.iloc[0:5, 0:5]

print("\n--- Top 5 Records of Heaviest Meteorites ---")
print(heaviest_meteorites)
```

```python
# Filtering: Apply multiple boolean conditions using & and | to find 
#         meteorites that are both greater than 50,000 grams in mass 
#         and were discovered in a specific region 
#         (e.g., "Antarctica" or "Sahara").

meteorite = df[
    (df['mass_grams'] > 50000) & 
    (df['region'].str.contains('Antarctica|Africa', na=False))
]

print("--- Heavy Meteorites from Antarctica or Africa ---")
print(meteorite)
```

```python
# Exporting Filtered Data

meteorite.to_excel('meteorite.xlsx', index=False)

meteorite.to_json('meteorite.json', orient='records')

print("File Successfully Exported")
```
