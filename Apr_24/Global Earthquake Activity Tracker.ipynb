```python
# The Scenario: You act as junior seismologists analyzing recent global 
#             earthquake data.
# Data I/O: You start by loading the data using pd.read_csv(). Later, you
#             will export their filtered findings using .to_excel() or
#             .to_json() for a web team to use.
# Data Structures: You separate out specific columns, like the magnitude, 
#             to see how it behaves as an isolated 1D Series compared to 
#             the 2D DataFrame context.
                
import pandas as pd

df = pd.read_csv('earthquake.csv')

print("2D Dataframe Context:")
print(df.head())

magnitude_series = df['mag']

print("\nIsolated 1D Series:")
print(magnitude_series.head())
```

```python
# Initial Data Inspection: Before doing anything, you must run .info() 
#             and .describe() to understand the sheer volume of data, 
#             use .head() to see the layout, and use .value_counts() 
#             to see which regions have the most frequent seismic activity.
print("--- Data Layout ---")
print(df.head())

print("\n--- Structural Summary ---")
df.info()

print("\n--- Statistical Distribution ---")
print(df.describe())

regions = df['place'].str.split(',').str[-1].str.strip()
region_counts = regions.value_counts()

print("\n--- Seismic Frequency by Region ---")
print(region_counts)
```

```python
# Indexing and Selection: 
# Column/Label/Integer: Use dictionary-style selection to grab the 
#             'Location' column. Use .iloc[] to extract the top 5 
#             largest earthquakes based on row position.
location_data = df['place']

print("--- Location Column ---")
print(location_data.head())

df_sorted = df.sort_values(by='mag', ascending=False)
largest_earthquakes = df_sorted.iloc[0:5, [4, 13]]
largest_earthquakes['place']  = largest_earthquakes['place'].str.split(',').str[-1].str.strip()

print("\n--- Top 5 Records of Largest Earthquarkes ---")
print(largest_earthquakes)
```

```python
# Filtering: Apply multiple boolean conditions using & and | to find 
#         earthquakes that are both greater than magnitude 6.0 and 
#         occurred in a specific region (e.g., "Japan" or “Texas”).
big_quakes = df[
    (df['mag'] > 6.0) & 
    (df['place'].str.contains('Alaska') | df['place'].str.contains('Indonesia'))
].copy()

print("--- Significant Events in Japan or Texas (> 6.0) ---")
big_quakes['place']  = big_quakes['place'].str.split(',').str[-1].str.strip()
print(big_quakes[['time', 'mag', 'place']])
```

```python
# Exporting Filtered Data
big_quakes.to_json('big_quakes.json', orient='records')

big_quakes.to_excel('big_quakes.xlsx', index=False)

print("File successfully exported")
```
