```python
# The Interactive Mad Libs Game
noun = input("Noun: ")
adj = input("Adjective: ")
verb = input("Verb: ")
print(f"""Today I found a {adj} {noun} in my kitchen.

Before I could react, it suddenly {verb}. 

We stared at each other for a full ten seconds.

Then it {verb} again… but this time louder."""); 
                
```

```python
# Restaurant Tip Calculator
bill = float(input("Bill amount: "))
tip_percent = float(input("Tip %: "))

tip_amount = bill * (tip_percent / 100)
total = bill + tip_amount

print(f"Tip:   ${tip_amount:.2f}")
print(f"Total: ${total:.2f}")
```

```python
# The Road Trip Cost Estimator
dist  = float(input("Total distance (km): "))
kpl   = float(input("What is your car's average KPL (km/l): "))
price = float(input("Fuel price per liter ($): "))

liters = dist / kpl
total_cost = liters * price

print(f"\nFuel needed: {liters:.2f} L")
print(f"Total Cost:  ${total_cost:.2f}")
```
