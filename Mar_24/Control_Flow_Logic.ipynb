```python
# Number Guessing Game
import random

target = random.randint(1, 100)
guess = 0 

while guess != target:
    guess = int(input("Guess (1-100): "))

    if guess > target:
        print("Too high!")
    elif guess < target:
        print("Too low!")
    else:
        print(f"You got it! The number was {target}.")

print("Game Over.")
```

```python
# Rock, Paper, Scissors
import random

options = ("rock", "paper", "scissors")
score = 0
active = True

while active:

    user = input("Choice (rock/paper/scissors/quit): ").lower()
    
    if user == "quit":
        active = False
        continue

    if user not in options:
        print(f"Invalid choice!")
        continue
        
    comp = random.choice(options)
    print(f"Computer chose: {comp}")

    if user == comp:
        print("It's a tie!")
    elif (user == "rock" and comp == "scissors") or \
         (user == "paper" and comp == "rock") or \
         (user == "scissors" and comp == "paper"):
        score += 1
        print("You win!")
    else:
        print("You lose!")
print(f"Total Score: {score}")
```

```python
# The Virtual Barista (Simple Point-of-Sale)
menu = {"Coffee": 3.0, "Tea": 2.5, "Latte": 4.5}
total = 0.0

print("--- MENU ---")
print("Coffee: $3.0 | Tea: $2.5 | Latte: $4.5")

while True:
    order = input("\nWhat would you like? ").lower()
    print("(Type 'checkout' to finish)")
    if order == "checkout":
        break

    if order == "coffee":
        total += menu["Coffee"]
    elif order == "tea":
        total += menu["Tea"]
    elif order == "latte":
        total += menu["Latte"]
    else:
        print("Sorry, that's not on the menu.")

print(f"Final total is: ${total:.2f}")
```

```python
# The Mental Math Challenger
import random

score = 0
answer = True

while answer:
    num1 = random.randint(1, 12)
    num2 = random.randint(1, 12)
    result = num1 * num2
    
    user_answer = int(input(f"What is {num1} x {num2}? "))
    
    if user_answer == result:
        score += 1
        print(f"Correct! Score: {score}")
    else:
        print(f"Wrong! The answer was {result}.")
        answer = False

print(f"Game Over. Final score: {score}")
```
