**Methods and Attributes**  
**Goal:** Distinguish between fitting, predicting, and accessing learned parameters.  
**Task:** Instantiate an SGDClassifier.
Deliverable: Call fit() on some dummy data. Then, extract and print the learned weights (coef_ and intercept_ attributes—note the trailing underscore, a key API design choice). Finally, use both predict() and decision_function(), and explain the difference between their outputs.
    

```python
import numpy as np
from sklearn.linear_model import SGDClassifier

# 1. Instantiate the SGDClassifier
sgd = SGDClassifier(loss='hinge', random_state=42)

# Create dummy data: 4 samples, 2 features
X = np.array([[1, 2], [2, 3], [4, 5], [5, 6]])
y = np.array([0, 0, 1, 1])

# 2. Fit the model on dummy data
sgd.fit(X, y)

# 3. Extract and print the learned parameters
# Trailing underscores indicate attributes learned during fitting
print("Learned Weights (coef_):", sgd.coef_)
print("Learned Bias (intercept_):", sgd.intercept_)
print("-" * 50)

# 4. Predict and compute decision function scores for a new sample
X_new = np.array([[3, 4]])
prediction = sgd.predict(X_new)
score = sgd.decision_function(X_new)

print("New sample:", X_new.tolist())
print("predict() output:", prediction)
print("decision_function() output:", score)
```
