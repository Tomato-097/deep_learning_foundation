**Dataset:** Synthetic data (manually typed lists).  
**Goal:** Reinforce the Matplotlib Object-Oriented canvas structure and introduce basic Seaborn styling to track how two different medications reduce pain over time.

### 1. Importing Libraries & Theming:
- Start by importing matplotlib.pyplot as plt and import seaborn as sns.  
- New Skill: Before creating your plot, use Seaborn to instantly upgrade the visuals by setting a theme. Call sns.set_theme(style="darkgrid").

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

### 2. The Anatomy / Basic Plotting:
- Create lists to track a 6-hour window.
- hours = [1, 2, 3, 4, 5, 6]
- Create two lists for the medications, representing reported pain levels on a scale of 0 to 10 (e.g., med_A_pain = [8, 6, 5, 3, 2, 2]).

```python
hours = [1, 2, 3, 4, 5, 6]

med_A_pain = [8, 6, 4, 2, 0, 0]
med_B_pain = [8, 7, 6, 4, 4, 2]
```

Set up your canvas using the Object-Oriented method: fig, ax = plt.subplots().  
Plot the data using ax.plot(hours, med_A_pain) and ax.plot(hours, med_B_pain).  
New Skill: Inside your ax.plot() calls, try adding markers and specific colors to differentiate the lines further (e.g., ax.plot(..., color='blue', marker='o', label='Medication A')).


```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(hours, med_A_pain, color='blue', marker='o')
ax.plot(hours, med_B_pain, color='green', marker='s')

plt.show()
```

### 3. Basic Customization:
Instruct the canvas to display your metadata:
- ax.set_title("Pain Level Over Time: Med A vs. Med B")
- ax.set_xlabel("Hours Post-Dose")
- ax.set_ylabel("Reported Pain Level (0-10)")  
Call ax.legend() so the viewer knows which line represents which medication.
Finally, render the plot cleanly with plt.show().

```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(hours, med_A_pain, color='blue', marker='o', label='Medication A')
ax.plot(hours, med_B_pain, color='green', marker='s', label='Medication B')

ax.set_title("Pain Level Over Time: Med A vs. Med B")
ax.set_xlabel("Hours Post-Dose")
ax.set_ylabel("Reported Pain Level (0-10)")
ax.legend()

plt.show()
```

### 4. Explain each of the graphs in details(both Khmer and English)

Over the course of this six-hour trial, we see two very different stories of patient recovery. Both patients began the morning in severe distress, starting at a high pain level of 8, which set a fair and equal baseline for both treatments. Medication A, represented by the blue line, immediately took control of the situation with a fast and perfectly steady pace. By dropping exactly two levels every hour, it acted like a high-performance machine, reaching the ultimate goal of zero pain by the fifth hour. The plateau at the end of the graph—where the line runs perfectly flat—is a huge win for the patient; it proves that the medicine didn't just work quickly, but it also stayed in the system long enough to keep the pain suppressed through the final hour.

In contrast, the green line for Medication B shows a much more frustrating and unsteady experience. While it started out with a slow but consistent decline, it showed a sudden burst of effectiveness between the third and fourth hour. However, this progress immediately stalled. Between the fourth and fifth hour, the graph shows another plateau, but this one is a sign of failure rather than success. Instead of continuing to drop, the medication hit a "dead zone" where the patient remained stuck in moderate pain for two full hours. While it eventually managed one final drop to level 2, the overall trend suggests the drug is unreliable and likely wearing off before the patient can reach total relief. As the blue line hits zero and stays there, it justifies its higher price by offering a guaranteed, lasting cure, while the green line leaves the patient in a state of lingering discomfort.
