```python
# Modular Unit Converter:
# The Goal: A program that converts temperatures (Celsius to Fahrenheit), 
# distances (Miles to Kilometers), and weights(Grams to Kilograms).
# The Catch: Every single conversion must be its own independent function 
# (e.g., def c_to_f(celsius):). There should be a main menu function that 
# calls the others based on user choice.
# Concepts Tested: Defining functions, passing arguments, return statements,
# variable scope.

def c_to_f(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def mi_to_km(miles):
    km = miles * 1.60934
    return km

def g_to_kg(grams):
    kg = grams / 1000
    return kg

def main_menu():
    while True:
        print("\n--- Modular Unit Converter ---")
        print("1. Celsius to Fahrenheit")
        print("2. Miles to Kilometers")
        print("3. Grams to Kilograms")
        print("4. Exit")
        
        choice = input("Select an option (1-4): ")
        
        if choice == "4":
            print("Goodbye!")
            break
            
        if choice in ["1", "2", "3"]:
           while True:
                try:
                    value = float(input("Enter the value to convert: "))
                    
                    if choice == "1":
                        result = c_to_f(value)
                        print(f"{value}°C is {result}°F")
                    elif choice == "2":
                        result = mi_to_km(value)
                        print(f"{value} miles is {result} km")
                    elif choice == "3":
                        result = g_to_kg(value)
                        print(f"{value} grams is {result} kg")
                    break
                except ValueError:
                    print("Error: Please input a valid number.")

main_menu()
```

```python
# Password Generator:
# The Goal: A function that takes an integer (length) and boolean flags 
# (use_numbers, use_symbols) and returns a randomized, secure password.
# Concepts Tested: Default arguments, keyword arguments, combining lists/strings, 
# returning data.

import random
import string

def gen_pass(length=12, use_numbers=True, use_symbols=False):
    char_pool = string.ascii_letters 
    
    if use_numbers:
        char_pool += string.digits 
        
    if use_symbols:
        char_pool += string.punctuation 
    
    password = ""
    
    for i in range(length):
        random_char = random.choice(char_pool)
        password += random_char
        
    return password

print(gen_pass())
print(gen_pass(use_numbers=False, use_symbols=True))
```

```python
# The Text Analyzer Suite
# The Goal: Ask the user to input a long sentence or paragraph. Build a 
# program that outputs statistics about their text.
# The Catch: You must use separate functions for every analysis: 
# count_words(text), count_vowels(text), and find_longest_word(text). 
# Each function must take the text as an argument and return the result to 
# be printed by the main program.
# Concepts Tested: String manipulation, passing strings as arguments, 
# returning specific values (integers and strings) from functions, and keeping 
# logic strictly separated.

def count_words(text):
    words = text.split()
    return len(words)

def count_vowels(text):
    vowels = "aeiou"
    count = 0
    for char in text:
        if char.lower() in vowels:
            count += 1
    return count

def longest_word(text):
    words = text.split()
    longest = words[0]
    for word in words:
        if len(word) > len(longest):
            longest = word
    return longest

user_text = input("Enter a long sentence or paragraph: ")

words = count_words(user_text)
vowels = count_vowels(user_text)
longest = longest_word(user_text)

print("\n--- Text Analysis Statistics ---")
print(f"Total Word: {words}")
print(f"Total Vowel: {vowels}")
print(f"Longest Word: '{longest}'")
```

```python
# The Virtual ATM (Banking Simulator)
# The Goal: Simulate a bank account where a user can deposit, withdraw, or 
# check their balance.
# The Catch: Create specific functions like deposit(balance, amount) and 
# withdraw(balance, amount). These functions must return the newly calculated 
# balance to overwrite the old balance in the main program loop. The withdraw 
# function must also include logic to prevent overdrafting (returning an error
# message and the unmodified balance if they ask for too much).
# Concepts Tested: Passing variables as arguments, updating state using return
# statements, handling edge cases within functions, and understanding variable
# scope (realizing that a variable modified inside a function doesn't change 
# outside unless returned).

def deposit(balance, amount):
    new_balance = balance + amount
    print(f"Successfully deposited ${amount:.2f}")
    return new_balance

def withdraw(balance, amount):
    if amount > balance:
        print("Error: Insufficient funds! Transaction cancelled.")
        return balance
    else:
        new_balance = balance - amount
        print(f"Successfully withdrew ${amount:.2f}")
        return new_balance

curr_b = 1000.00  
while True:
    print(f"\n--- Virtual ATM (Current Balance: ${curr_b:.2f}) ---")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")
        
    choice = input("Select an option: ")
    
    if choice == "4":
        print("Thank you for using Virtual ATM. Goodbye!")
        break
        
    if choice == "1":
        print(f"Your current balance is: ${curr_b:.2f}")   
        
    elif choice == "2" or choice == "3":
        try:
            amt = float(input("Enter amount: "))
            if amt <= 0:
                print("Error: Amount must be a positive number.")
                continue
            if choice == "2":
                curr_b = deposit(curr_b, amt)
            else:
                curr_b = withdraw(curr_b, amt)
        except ValueError:
            print("Error: Invalid input. Please enter a numeric value.")
    else:
        print("Invalid choice.")
```
