```python
#### Memory-Based To-Do List:
# The Goal: An application that lets a user add a task, view all tasks, 
# remove a task, and exit. Store the tasks in a Python List.
# Concepts Tested: List methods (.append(), .pop()), indexing, 
# iterating through lists with for loops.

tasks = []

while True:
    print("\n1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")
    
    choice = input("Select an option: ")

    if choice == "1":
        new_task = input("Add task: ").strip()
        if new_task:
            tasks.append(new_task)
            print("Task added.")
        else:
            print("Error: Task cannot be empty or just spaces.")
    
    elif choice == "2":
        print("\nCurrent Tasks:")
        if not tasks:
            print("The list is empty.")
        for i in range(len(tasks)):
            print(f"{i + 1} - {tasks[i]}")
            
    elif choice == "3":
        if not tasks:
            print("The list is empty.")
        else:
            print("\nCurrent Tasks:")
            print("Index - Tasks")
            for i in range(len(tasks)):
                print(f"{i}-{tasks[i]}")
            
            removal = input("Enter index or task to remove: ").strip()
            
            if removal.isdigit():
                index = int(removal)
                if 0 <= index < len(tasks):
                    removed = tasks.pop(index)
                    print(f"Task '{removed}' removed.")
                else:
                    print("Error: That index does not exist.")
            elif removal in tasks:
                tasks.remove(removal)
                print(f"Task '{removal}' removed.")
            else:
                print("Error: Task name not found.")
        
    elif choice == "4":
        print("Exiting...")
        break
```

```python
# Interactive Flashcard / Quiz App:
# The Goal: Create a dictionary where the keys are questions(e.g., countries) 
# and values are answers (e.g., capitals). Loop through the dictionary, 
# ask the user the question, check their answer, and calculate a final score.
# Concepts Tested: Dictionaries, key-value pairs, iteration, score tracking.

dictionary = {
    "France": "Paris",
    "Japan": "Tokyo",
    "Germany": "Berlin",
    "Italy": "Rome"
}

score = 0

for country, capital in dictionary.items():
    user_ans = input(f"What is the capital of {country}? ")
    if user_ans.lower() == capital.lower():
        print("Correct!")
        score += 1
    else:
        print(f"Wrong! The correct answer is {capital}.")

print(f"Final score is: {score}/{len(dictionary)}")
```

```python
# The Shopping Cart Analyzer
# The Goal: Ask the user to input the prices of items they are adding to their 
# cart, one by one. Store these prices in a Python List. The loop stops when 
# the user types 'done'. Then, calculate and print a receipt showing the total 
# cost, the most expensive item in the cart, and the average item price.
# Concepts Tested: while loops for input collection, list methods (.append()), 
# built-in list functions (sum(), max(), len()), and data type casting 
# (handling the mix of string commands and float prices).

prices = []

while True:
    user_input = input("Input item  price (or done): ")
    if user_input.lower() == 'done':
        break
    try:
        price = float(user_input)
        prices.append(price)
    except ValueError:
        print("Error. Please enter a numeric price or 'done'.")
if len(prices) > 0:
    total = sum(prices)
    most = max(prices)
    ave = total / len(prices)

    print("\n--- Receipt ---")
    print(f"Total Cost: {total}")
    print(f"Most Expensive Item: {most}")
    print(f"Average Price: {ave}")
else:
    print("No items in the cart.")
```

```python
# The Coffee Shop Inventory Manager
# The Goal: Create a dictionary pre-filled with cafe supplies and their quantities 
# (e.g., {'coffee beans': 50, 'cups': 100, 'syrup': 15}). Create a loop that asks 
# the user if they want to "sell" or "restock" an item. Update the dictionary 
# values based on their input and print the current inventory levels. 
# Prevent the user from selling more items than they have in stock!
# Concepts Tested: Accessing and updating dictionary values, checking if a key
# exists using the in operator, combining if/elif logic with dictionary lookups, 
# and managing state.

inventory = {
    "coffee beans": 50, 
    "cups": 100, 
    "syrup": 15
}

while True:
    print("\n--- Current Inventory ---")
    for item, quantity in inventory.items():
        print(f"{item}: {quantity}")

    action = input("\nDo you want to 'sell', 'restock', or 'exit'? ").lower()

    if action == "exit":
        print("Exiting Inventory Manager. Goodbye!")
        print("-" * 25)
        break

    if action == "sell" or action == "restock":
        target_item = input("Enter the item name: ").lower()

        if target_item in inventory:
            amount = int(input(f"Enter amount to {action}: "))

            if action == "restock":
                inventory[target_item] += amount
                print(f"Successfully restocked {target_item}.")

            elif action == "sell":
                if amount <= inventory[target_item]:
                    inventory[target_item] -= amount
                    print(f"Successfully sold {amount} {target_item}.")
                else:
                    print(f"Not enough stock! You only have {inventory[target_item]}.")
        else:
            print("That item does not exist in our records!")
    else:
        print("Error. Please type 'sell', 'restock', or 'exit'.")
```
