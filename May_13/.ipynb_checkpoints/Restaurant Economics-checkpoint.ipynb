**Dataset:** sns.load_dataset('tips') 

**Goal:** Find out what factors drive higher tips at a restaurant.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
```

**Num vs. Num - Scatter:**  
Create a sns.scatterplot() with total_bill on the X-axis and tip on the Y-axis to prove that higher bills equal higher tips.

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs. Tip')
plt.show()
```

**Num vs. Num - Line:**  
Create a mock timeline of total_bill values using sns.lineplot() to practice connecting data points.

```python
sns.lineplot(data=tips, x='total_bill', y='tip', marker='o')

plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Mock Timeline of Total Bill Values')
plt.show()
```

**Num vs. Cat - Box Plots:**  
Use sns.boxplot() with day on the X-axis and total_bill on the Y-axis to find the median bill and outliers for each day of the week.

```python
sns.boxplot(data=tips, x='day', y='total_bill')

plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.title('Total Bill Distribution by Day')
plt.show()
```

**Num vs. Cat - Violin Plots:**  
Use sns.violinplot() comparing time (Lunch vs. Dinner) and tip to see where the density of generous tippers lies.

```python
sns.violinplot(data=tips, x='time', y='tip')

plt.xlabel('Time of Day')
plt.ylabel('Tip ($)')
plt.title('Density of Tip Amounts by Time of Day')
plt.show()
```

**Num vs. Cat - Bar Plots:**  
Use sns.barplot() to show the average tip amount grouped by size (party size), letting Seaborn calculate the mean automatically.

```python
sns.barplot(data=tips, x='size', y='tip')

plt.xlabel('Party Size')
plt.ylabel('Average Tip ($)')
plt.title('Average Tip Amount by Party Size')
plt.show()
```
